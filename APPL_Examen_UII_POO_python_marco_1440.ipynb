## clase CANCION
Crear la clase "Canción", con 3 atributos (titulo, artista, album), 2 funciones leerDatos(), mostrarDatos() y 2 objetos (cancion1, cancion2).
![texto del vínculo](https://raw.githubusercontent.com/GabrielMorquecho128/ExaUII-Clase-cancion-marco-1440/refs/heads/main/InfoexaIIclasscancion.png)

```python
# ExaUII clase Canción Marco 1440
# Identificación: Marco Morquecho | ID: 1440 | NL: 39
# Descripción: Implementación interactiva de la Clase Canción con 3 atributos y 2 métodos, demostrada con 2 objetos.

# --- DECLARACIÓN DE LA CLASE ---
class Cancion:
    # Atributos de instancia (se definen en leerDatos)

    # Método 1: Leer datos desde el teclado (Interactividad V3)
    def leerDatos(self):
        print("\n--- Registro de Canción ExaUII ---")
        self.titulo = input("Ingresa el título de la canción: ")
        self.artista = input("Ingresa el artista/banda: ")
        self.album = input("Ingresa el álbum: ")

    # Método 2: Mostrar datos en forma de Reporte
    def mostrarDatos(self):
        print("\n" + "="*40)
        print(f"FICHA TÉCNICA: {self.titulo}")
        print(f"Artista: {self.artista}")
        print(f"Álbum:   {self.album}")
        print("="*40)

# --- PROGRAMA PRINCIPAL (Instanciación) ---
print("SISTEMA DE GESTIÓN MUSICAL INTERACTIVA V3")

# Creación y llenado del Objeto 1
cancion1 = Cancion()
print("\nDATOS PARA LA CANCIÓN 1:")
cancion1.leerDatos()

# Creación y llenado del Objeto 2
cancion2 = Cancion()
print("\nDATOS PARA LA CANCIÓN 2:")
cancion2.leerDatos()

# --- RESULTADOS FINALES ---
print("\n=== RESUMEN DE FLOTA MUSICAL REGISTRADA ===")
cancion1.mostrarDatos()
cancion2.mostrarDatos()

print("\n" + "="*40)
print("Programa realizado por Marco Morquecho NL=39 | ID=1440")
print("="*40)
```
